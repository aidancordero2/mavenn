## One-hot encode the data ... this takes a few minutes

```python
temp_x_train = []
temp_x_test = []

for reshape_index in range(len(x_train)):
    temp_x_train.append(list(np.array(x_train)[reshape_index]))

for reshape_test_index in range(len(x_test)):
    temp_x_test.append(list(np.array(x_test)[reshape_test_index]))
    
x_train = temp_x_train
x_test = temp_x_test

x_train = np.array(x_train)
x_test = np.array(x_test)
```

```python
base_dict = {"K":0, "R":1, "H":2, "E":3, "D":4, "N":5, "Q":6, "T":7, "S":8, "C":9, "G":10, "A":11, "V":12, "L":13, "I":14, "M":15, "P":16, "Y":17, "F":18, "W":19}
```

```python
test_size = len(y_test)
```

```python
x_train_tensor = np.zeros(list(x_train.shape) + [20])    # shape: (batch_size, 4)
x_test_tensor = np.zeros(list(x_test.shape) + [20])    # shape: (batch_size, 4)
base_dict = {"K":0, "R":1, "H":2, "E":3, "D":4, "N":5, "Q":6, "T":7, "S":8, "C":9, "G":10, "A":11, "V":12, "L":13, "I":14, "M":15, "P":16, "Y":17, "F":18, "W":19}

#base_dict = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
num_sample = len(x_train_tensor)
seq_length = len(x_train[0])
test_size = len(x_test)


#naive one-hot encoding
for row in range(num_sample):
    for col in range(seq_length):
        x_train_tensor[row,col,base_dict[x_train[row,col]]] = 1
        if(row<test_size):
            x_test_tensor[row,col,base_dict[x_test[row,col]]] = 1
```

### Calculate mutual information

```python
from sklearn.metrics import mutual_info_score
import numpy as np

def calc_MI(x, y, bins):
    c_xy = np.histogram2d(x, y, bins)[0]
    mi = mutual_info_score(None, None, contingency=c_xy)
    return mi
```

```python
# weights from the 1st dense layer of the optimized neural network
NN_weights = np.loadtxt('NN_dense_layer_weights.csv')
NN_weights = NN_weights.reshape(80,1)
```

### All this will go in another loop

```python
# initialize a set of weights with shape equal shape of x_train_tensor[0]
weights = np.random.uniform(low=0.00001, high=0.5, size=(x_train_tensor[0].flatten().shape[0],1))
#weights = NN_weights.copy()
bias = np.random.rand()

# the following computation for all training data
# then you'll have 80,000 predictions and 80,000 labels, 
# form histograms and compute the mutual information between them.

# number bins to use in mi computations
number_bins_mi = 100
accepted_mc_steps = 0
final_predictions_list = []
# helpful list to visualize increase in mi per step
mi_increase_list = []
mi_vs_time_list = []

# initialize predictions list
initialize_predictions_list = []
# loop over training data and compute predictions (y=wx+b)
initial_weights = weights.copy()
for initial_predictions_index in range(len(x_train_tensor)):
    initialize_predictions_list.append(x_train_tensor[initial_predictions_index].flatten().dot(initial_weights)[0]+bias)


for mc_index in range(50000):

    # will contain the set of predictions per step
    predictions_list = []
    intermediate_predictions_list = []

    # loop over training data and compute predictions (y=wx+b)
    for train_index in range(len(x_train_tensor)):
        predictions_list.append(x_train_tensor[train_index].flatten().dot(weights)[0]+bias)

    # compute current mutual information
    current_mi = calc_MI(predictions_list,y_train.flatten(),number_bins_mi)    
    #print(current_mi)
    # Select weights randomly
    number_weights_to_change = np.random.randint(x_train_tensor[0].flatten().shape[0])
    random_weight_indices = np.random.randint(x_train_tensor[0].flatten().shape[0], size=(number_weights_to_change,))

    # Make small change to weights
    learning_parameter = 0.01
    intermediate_weights = weights.copy()
    update_to_weights = np.random.uniform(-1,1,size=(number_weights_to_change,1))
    intermediate_weights[random_weight_indices]+= learning_parameter*update_to_weights
    
    intermediate_bias = np.random.rand()
    # compute updated predictions
    for train_index in range(len(x_train_tensor)):
        intermediate_predictions_list.append(x_train_tensor[train_index].flatten().dot(intermediate_weights)[0]+bias)

    # compute intermediate mutual information
    intermediate_mi = calc_MI(intermediate_predictions_list,y_train.flatten(),number_bins_mi)
    
    if(mc_index%100==0):
        print(mc_index,accepted_mc_steps, current_mi)
    
    # updates if move accepted
    if((intermediate_mi-current_mi)>0):
        weights = intermediate_weights
        bias = intermediate_bias
        #print(mc_index,accepted_mc_steps, intermediate_mi)
        current_mi = intermediate_mi
        mi_increase_list.append(intermediate_mi-current_mi)
        accepted_mc_steps+=1
        mi_vs_time_list.append(current_mi)
        final_predictions_list = intermediate_predictions_list.copy()

print('done')
    
```

```python
mi_vs_time_list
figure_size = [6,5]
plt.figure(figsize=figure_size)
plt.tick_params(labelsize=14)
plt.xlabel('Accepted Steps',fontsize=16)
plt.ylabel('MI',fontsize=16)
plt.plot(mi_vs_time_list,'o',color='black')
plt.show()
```
