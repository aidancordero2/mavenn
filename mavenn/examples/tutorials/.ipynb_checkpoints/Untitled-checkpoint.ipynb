```python
# Standard imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# For showing images
from IPython.display import Image

# Insert path to mavenn beginning of path
import sys
sys.path.insert(0, '../../../')

# Load mavenn
import mavenn
```

```python
cons_ss = "CAGGTAAGT"
model = mavenn.load_example_model('mpsa_ge_additive')
```

```python
df1 = mavenn.get_1pt_effects(func=model.x_to_phi, alphabet='dna', wt_seq=cons_ss)
df1.head()
```

```python
df2 = mavenn.get_2pt_effects(func=model.x_to_phi, alphabet='dna', wt_seq=cons_ss)
df2 = df2.drop(columns=["dphi"])
df2 = df2.rename(columns={'phi':'phi_12'})
df2 = df2.iloc[1:,:]
mut_names = np.array([name.split(',') for name in df2.index])
df2['mut1'] = mut_names[:,0]
df2['mut2'] = mut_names[:,1]

df2 = df2.merge(left_on="mut1", right_index=True, right=df1["phi"], how='left')
df2 = df2.rename(columns={'phi':'phi_1'})

df2 = df2.merge(left_on="mut2", right_index=True, right=df1["phi"], how='left')
df2 = df2.rename(columns={'phi':'phi_2'})
df2["phi_wt"] = df1.loc['WT','phi']

df2 = df2[["l1", "c1_wt", "c1_mut", 
           "l2", "c2_wt", "c2_mut", 
           "phi_12", "phi_1", "phi_2", "phi_wt",]]
df2["ddphi"] = df2["phi_12"] - df2["phi_1"] - df2["phi_2"] + df1.loc['WT','phi']
df2.head()
```
